```python
import pandas as pd #imported pandas
```

```python
#data
#one column is we called as Series
#combination of series are DataFrame
data = {
   'ages': [14, 18, 24, 42],
   'heights': [165, 180, 176, 184]
}
```

```python
df = pd.DataFrame(data) #created dataFrame from 2 series

#print it
print(df)
```

The DataFrame automatically creates a numeric index for each row.
**We can specify a custom index, when creating the DataFrame âœ¨**

```python
df = pd.DataFrame(data, index=['James', 'Bob', 'Amy', 'Dave'])
print(df)
```

```python
#we can access a row using its index and the loc[] function
print(df.loc["Amy"])
```

```python
#print specific column/Series
print(df["ages"])
```

```python
#Pandas uses the iloc function to select data based on its numeric index.
x = df.iloc[2]
print(x)
```
