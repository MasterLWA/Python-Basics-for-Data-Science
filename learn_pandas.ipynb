```python
import pandas as pd #imported pandas
```

```python
#data
#one column is we called as Series
#combination of series are DataFrame
data = {
   'ages': [14, 18, 24, 42],
   'heights': [165, 180, 176, 184]
}
```

```python
df = pd.DataFrame(data) #created dataFrame from 2 series

#print it
print(df)
```

The DataFrame automatically creates a numeric index for each row.
**We can specify a custom index, when creating the DataFrame âœ¨**

```python
df = pd.DataFrame(data, index=['James', 'Bob', 'Amy', 'Dave'])
print(df)
```

```python
#we can access a row using its index and the loc[] function
print(df.loc["Amy"])
```

```python
#print specific column/Series
print(df["ages"])
```

```python
#Pandas uses the iloc function to select data based on its numeric index.
x = df.iloc[2]
print(x)
```

# Reading data
* It is quite common for data to come in a file format. One of the most popular
formats is the CSV (comma-separated values).
* read_csv() function reads the data of a CSV file into a DataFrame


```python
df = pd.read_csv("https://www.sololearn.com/uploads/ca-covid.csv")
```

* We can get the first rows of the data using the head() function
* **By default it returns the first 5 rows**


```python
df.head()
```

can instruct it to return the number of rows you would like as an argument

```python
df.head(20)
```

* Print last rows of the dataframe

```python
df.tail(10) #get last 10 rows
```

* The info() function is used to get information about our dataset

```python
df.info()
```

* We can set our own index column by using the set_index()
* The inplace=True argument specifies that the change will be applied to our DataFrame, without the need to assign it to a new DataFrame variable.

```python
# df.set_index('Date')
# print(df.head())
```

* drop() deletes rows and columns.
* axis=1 specifies that we want to drop a column.
* axis=0 will drop a row.

```python
df.drop('state', axis=1, inplace=True)
df.head()
```

* add new column to dataframe

```python
df['month'] = pd.to_datetime(df['date'], format="%d.%m.%y").dt.month_name()
df.head()
```

* value_counts() returns how many times a value appears in the dataset, also called the frequency of the values.

```python
df.value_counts('month')
```

* The groupby() function is used to group our dataset by the given column.
* Similarly, we can use min(), max(), mean(), etc. to find the corresponding values for each group.



```python
df.groupby('month')['cases'].sum()
```
